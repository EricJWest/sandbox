```python
# import packages
import numpy as np
%matplotlib notebook
import matplotlib.pyplot as plt
import matplotlib.animation as ani
import time
```

```python
# Set up the figure, axis, and plot element to animate
fig = plt.figure()
ax = plt.axes(xlim=(0, 2), ylim=(-2, 2))
line, = ax.plot([], [], lw=2)
```

```python
# soliton function
def soliton(t,x):
    return 0.5/np.cosh(0.5*(x - t))**2
```

```python
# xgrid
x = np.linspace(0,60.0,1001)
```

```python
# time
tInitial = 0   # start time
tFinal = 60.0  # end time
dt = 0.1       # time increments
```

```python
# plot
plt.xlabel('x')
plt.ylabel('y')
line, = plt.plot(x, soliton(0,x))
plt.show()
```

```python
for t in np.arange(tInitial, tFinal, dt):
    line.set_ydata(soliton(t,x))
    plt.draw()
    plt.title('Soliton wave at t = %5.1f' % t)
    time.sleep(0.01)
```

```python
def draw_frame(t,x):
    """Draw a frame. """
    plt.plot(x, soliton(t,x))
    plt.axis([0,60.0,0,0.5])
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Soliton wave at t = %5.1f' % t)

x = np.linspace(0, 60.0, 1001)
t = np.linspace(-10, 70, 50)

for i in range(len(t)):
    file_name='_temp%05d.png' % i
    draw_frame(t[i],x)
    plt.savefig(file_name)
    plt.clf()

import os
os.system("rm soliton.avi")
os.system("ffmpeg -r 25 " +
        " -i _temp%05d.png -b:v 1800 soliton.avi")
os.system("rm _temp*.png")
```
